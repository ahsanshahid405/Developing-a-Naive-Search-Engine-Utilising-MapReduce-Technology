Removing the unnecessary Columns

```python
sample.dropna(inplace=True)
```

```python
columns_to_remove = ['TITLE', 'SECTION_TITLE']
sample = sample.drop(columns=columns_to_remove)
sample
```

Grouping the data

```python
# Grouping the data by 'ARTICLE_ID' as there are many article_ids with the same indexes for different SECTION_TITLE
sample = sample.groupby('ARTICLE_ID')['SECTION_TEXT'].apply(lambda x: ' '.join(x)).reset_index(name='SECTION_TEXT')
sample
```

```python
sample['SECTION_TEXT'].fillna(' ', inplace=True)
```

Convert textual data into lower case 

```python
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].str.lower()
```

```python
sample
```

```python
def remove_punctuation_manual(text):
    punctuation_chars = '''!()-[]{};:'"\,<>./|=?@#$%^+&*_~'''
    # Remove punctuation characters from the text
    text_without_punctuation = ''.join(char for char in text if char not in punctuation_chars)
    return text_without_punctuation

sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_punctuation_manual)
sample
```

```python
sample
```

```python
def remove_extra_line(text):
    punctuation_chars = '\n'
    # Remove punctuation characters from the text
    text_without_special_char = ''.join(char for char in text if char not in punctuation_chars)
    return text_without_special_char

sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_extra_line)
sample
```

```python
def tokenize_text(text):
    return word_tokenize(text)

# Stop-word removal function
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word.lower() not in stop_words]

# Tokenize and remove stopwords for each row
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(tokenize_text)
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_stopwords)
sample
```

```python
def create_vocabulary(section_text):
    vocabulary = set()
    for each_row in section_text:
        vocabulary.update(each_row)
    return vocabulary

# Create vocabulary from tokenized text in 'section_text' column
vocabulary = sorted(create_vocabulary(sample['SECTION_TEXT']))
```

Since vocabulary contains numeric values. I have not remove them because it may represent something important like code, any date or any id_of_event 

```python
print(vocabulary)
```

```python
# Use a counter to keep track of unique IDs
unique_id = 0

# Create a dictionary to store term-ID mapping
vocab_id = {}

# Assign unique IDs to each term
for word in vocabulary:
  vocab_id[word] = unique_id
  unique_id += 1  # Increment counter for next term

# Print the unique id for the vocabulary
print("The unique id for the vocabulary")
print(vocab_id)
```

Term frequency refers to the number of times a specific term or word appears in a document or a corpus of documents. It is a fundamental concept in natural language processing (NLP) and information retrieval.



```python
def calculate_tf(tokenized_text_column, vocab_id):
    term_freq = []
    for article_by_row in tokenized_text_column:
        term_frequency = {}
        for term in article_by_row:
            unique_word_id = vocab_id.get(term)
            if unique_word_id is not None:
                if unique_word_id in term_frequency:
                    term_frequency[unique_word_id] += 1
                else:
                    term_frequency[unique_word_id] = 1
        term_freq.append(term_frequency)
    return term_freq

# Calculate Term Frequency (TF) for each document
term_frequency = calculate_tf(sample['SECTION_TEXT'], vocab_id)
```

```python
print("Term Frequency")
for id, freq in enumerate(term_frequency):
    print(f"Document {id+1}: {freq}")
```

Inverse Document Frequency (IDF) is a measure used in information retrieval and text mining to quantify how important a term is within a collection of documents. It is based on the idea that terms that occur frequently across many documents are less informative than terms that occur rarely.

```python
def calculate_idf_for_vocab(section_text, vocab_id):
    idf_dict = {}
    N = len(section_text)
    for term, term_id in vocab_id.items():
        df_t = 0
        for row in section_text:
            if term_id in [vocab_id.get(word) for word in row]:
                df_t += 1
        idf_dict[term_id] = df_t
    return idf_dict

idf_values = calculate_idf_for_vocab(sample['SECTION_TEXT'], vocab_id)

print("IDF values:")
print(idf_values)
```

The TF-IDF weight indicates how important a term is to a document within a collection of documents. It gives higher weight to terms that are frequent within the document but rare in the corpus, effectively capturing the distinctive characteristics of the document.

```python
# Calculate TF/IDF weights
tfidf_weights = {}

# Total number of documents
total_documents = len(term_frequency)

# Calculate TF/IDF weights for each term in each document
for article_id, word_frequencies in enumerate(term_frequency):
    tfidf_weights[article_id] = {}
    total_words_in_document = sum(word_frequencies.values())
    for word_id, frequency in word_frequencies.items():
        # Calculate TF/IDF weight
        tf = frequency / total_words_in_document
        idf = idf_values.get(word_id, 0)
        tfidf_weights[article_id][word_id] = tf * idf


print("TF/IDF weights")
for article_id, weights in tfidf_weights.items():
    print(article_id + 1, end=" ")
    print(", ".join(f"({word_id}, {weight:.2f})" for word_id, weight in weights.items()))

```

TF-IDF to the Vector Space Model

```python
# Extracting the list of unique id's from the vocabulary where unique id is assigned
unique_word_ids = list(vocab_id.values())
print(unique_word_ids)
```

```python
# Define function to convert TF-IDF weights into vectors
def weights_to_vector(weights, vocabulary_size):
    vector = [0] * vocabulary_size
    for word_id, weight in weights.items():
        vector[word_id] = weight
    return vector

# Calculate TF/IDF weights
tfidf_weights = {}

# Total number of documents
total_documents = len(term_frequency)

# Calculate TF/IDF weights for each term in each document
for article_id, word_frequencies in enumerate(term_frequency):
    tfidf_weights[article_id] = {}
    total_words_in_document = sum(word_frequencies.values())
    for word_id, frequency in word_frequencies.items():
        # Calculate TF/IDF weight
        tf = frequency / total_words_in_document
        idf = idf_values.get(word_id, 0)
        tfidf_weights[article_id][word_id] = tf * idf

# Convert TF/IDF weights to vectors for each document
document_vectors = {}
for article_id, weights in tfidf_weights.items():
    vector = weights_to_vector(weights, len(unique_word_ids))
    document_vectors[article_id + 1] = vector

# Print document vectors
print("Document Vectors:")
for article_id, vector in document_vectors.items():
    print(f"Document {article_id}: {vector}")
```
