Data-Preprocessing

```python
sample = dataframe.read_csv('/home/ahsan/Documents/sample_data.csv')
```

```python
sample.info()
```

```python
sample
```

Removing the unnecessary Columns

```python
sample.dropna(inplace=True)
```

```python
columns_to_remove = ['TITLE', 'SECTION_TITLE']
sample = sample.drop(columns=columns_to_remove)
sample
```

Grouping the data

```python
# Grouping the data by 'ARTICLE_ID' as there are many article_ids with the same indexes for different SECTION_TITLE
sample = sample.groupby('ARTICLE_ID')['SECTION_TEXT'].apply(lambda x: ' '.join(x)).reset_index(name='SECTION_TEXT')
sample
```

```python
sample['SECTION_TEXT'].fillna(' ', inplace=True)
```

Convert textual data into lower case 

```python
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].str.lower()
```

```python
sample
```

```python
def remove_punctuation_manual(text):
    punctuation_chars = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    # Remove punctuation characters from the text
    text_without_punctuation = ''.join(char for char in text if char not in punctuation_chars)
    return text_without_punctuation

sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_punctuation_manual)
sample
```

```python
def remove_punctuation_manual(text):
    punctuation_chars = '\n'
    # Remove punctuation characters from the text
    text_without_punctuation = ''.join(char for char in text if char not in punctuation_chars)
    return text_without_punctuation

sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_punctuation_manual)
sample
```
