Data-Preprocessing

```python
sample = dataframe.read_csv('/home/ahsan/Documents/sample_data.csv')
```

Removing the unnecessary Columns

```python
sample.dropna(inplace=True)
```

```python
columns_to_remove = ['TITLE', 'SECTION_TITLE']
sample = sample.drop(columns=columns_to_remove)
sample
```

Grouping the data

```python
# Grouping the data by 'ARTICLE_ID' as there are many article_ids with the same indexes for different SECTION_TITLE
sample = sample.groupby('ARTICLE_ID')['SECTION_TEXT'].apply(lambda x: ' '.join(x)).reset_index(name='SECTION_TEXT')
sample
```

```python
sample['SECTION_TEXT'].fillna(' ', inplace=True)
```

Convert textual data into lower case 

```python
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].str.lower()
```

```python
sample
```

```python
def remove_punctuation_manual(text):
    punctuation_chars = '''!()-[]{};:'"\,<>./|=?@#$%^+&*_~'''
    # Remove punctuation characters from the text
    text_without_punctuation = ''.join(char for char in text if char not in punctuation_chars)
    return text_without_punctuation

sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_punctuation_manual)
sample
```

```python
sample
```

```python
def remove_extra_line(text):
    punctuation_chars = '\n'
    # Remove punctuation characters from the text
    text_without_special_char = ''.join(char for char in text if char not in punctuation_chars)
    return text_without_special_char

sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_extra_line)
sample
```

```python
def tokenize_text(text):
    return word_tokenize(text)

# Stop-word removal function
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word.lower() not in stop_words]

# Tokenize and remove stopwords for each row
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(tokenize_text)
sample['SECTION_TEXT'] = sample['SECTION_TEXT'].apply(remove_stopwords)
sample
```

```python
def create_vocabulary(section_text):
    vocabulary = set()
    for each_row in section_text:
        vocabulary.update(each_row)
    return vocabulary

# Create vocabulary from tokenized text in 'section_text' column
vocabulary = sorted(create_vocabulary(sample['SECTION_TEXT']))
```

Since vocabulary contains numeric values. I have not remove them because it may represent something important like code, any date or any id_of_event 

```python
print(vocabulary)
```

```python
# Use a counter to keep track of unique IDs
unique_id = 0

# Create a dictionary to store term-ID mapping
vocab_id = {}

# Assign unique IDs to each term
for word in vocabulary:
  vocab_id[word] = unique_id
  unique_id += 1  # Increment counter for next term

# Print the term-ID mapping
print("Term-ID Mapping")
print(vocab_id)
```
